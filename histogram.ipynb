```python
#원본 데이터 histogram
import matplotlib.pyplot as plt
import numpy as np
import cv2

img=cv2.imread('Kids_park.jpeg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

fig, ax = plt.subplots()
ax.hist(img.ravel(), 256, [0,256])
ax.set(title='Raw Data histogram',xlabel='bins', ylabel='count')

fig.savefig('Kids_park_hist.jpg')
```

```python
for i in range(2,10):
    img=cv2.imread('sp0{}.jpg'.format(i))
    img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    fig, ax = plt.subplots()
    ax.hist(img.ravel(), 256, [0,256])
    ax.set(title='S&P {}0% histogram'.format(i),xlabel='bins', ylabel='count')
    fig.savefig('sp0{}_hist.jpg'.format(i))
```
